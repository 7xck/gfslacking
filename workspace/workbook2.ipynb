```python
import random
import matplotlib.pyplot as plt

# Initialize variables
midprice = 1
spread = 0.001

inventory = 0
balance = 1000

fill_probability = 0.2


midprice_arr = []
inventory_arr = []
balance_arr = []

for _ in range(100000):
    price_movement = random.choice([-1, 1]) * 0.001
    midprice += price_movement

    if random.random() < fill_probability and inventory < 100:
        inventory += 1
        balance += spread
    if random.random() < fill_probability and inventory > -100:
        inventory -= 1
        balance += spread

    balance += inventory*price_movement

    midprice_arr.append(midprice)
    inventory_arr.append(inventory)
    balance_arr.append(balance)
    

figs, axis = plt.subplots(3)

axis[0].plot(midprice_arr, label="Midprice")
axis[1].plot(inventory_arr, label="Inventory")
axis[2].plot(balance_arr, label="Balance")
axis[0].legend()
axis[1].legend()
axis[2].legend()
plt.show()
```
